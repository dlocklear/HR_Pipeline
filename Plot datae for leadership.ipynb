```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the transformed data
file_path = 'data/transformed_Employee_data.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
print(df.head())

# Plot the distribution of salaries
plt.figure(figsize=(10, 6))
sns.histplot(df['Salary'], kde=True)
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()

# Plot the tenure of employees
plt.figure(figsize=(10, 6))
sns.histplot(df['tenure'], kde=True)
plt.title('Distribution of Employee Tenure')
plt.xlabel('Tenure (days)')
plt.ylabel('Frequency')
plt.show()

# Plot the average salary by position
plt.figure(figsize=(10, 6))
sns.barplot(x='Position', y='Salary', data=df, estimator=lambda x: sum(x) / len(x))
plt.title('Average Salary by Position')
plt.xlabel('Position')
plt.ylabel('Average Salary')
plt.show()
```
